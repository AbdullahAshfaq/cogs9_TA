## 2. Using API (Medium)

API call is similar to function call i.e. log(x)

Data is often returned in JSON format and has a defined structure

```python
import requests
import json
```

```python
# Getting random fact about cats
# https://github.com/wh-iterabb-it/meowfacts

resp = requests.get(url="https://meowfacts.herokuapp.com/")
resp = resp.content.decode('utf-8')
resp = json.loads(resp)
resp
```

```python
# Reminder: Getting fact from the JSON
# A. resp[0][0]
# B. resp['data'][0]
```

```python
# API call with parameters

# Getting random fact from 
resp = requests.get(url="https://meowfacts.herokuapp.com/?count=3")
resp = resp.content.decode('utf-8')
resp = json.loads(resp)
resp
```
